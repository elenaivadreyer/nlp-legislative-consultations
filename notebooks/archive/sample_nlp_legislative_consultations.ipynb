```python
import re

def extract_gesetz_structure(raw_text):
    """
    Extracts Articles and their sub-structures (Paragraphs §, Subparagraphs (1),
    or numbered Sections 1.) based on the structure typical for this law.
    """

    # -----------------------
    # 1. NORMALIZE LINE BREAKS & WHITESPACE
    # -----------------------
    text = raw_text.replace("\r", "")
    text = re.sub(r"[ \t]+\n", "\n", text)
    text = re.sub(r"\n[ \t]+", "\n", text)
    text = text.strip()

    # -----------------------
    # 2. FIND START AND SPLIT BY ARTIKEL
    # -----------------------
    # Sucht nach dem Start des ersten Artikel 1 Blockes
    start_match = re.search(r"^\s*Artikel\s+1\b", text, flags=re.MULTILINE | re.DOTALL)
    if not start_match:
         raise ValueError("Artikel 1 start not found.")
    text = text[start_match.start():]

    # Trennt den Text an jedem Artikel-Header (Artikel 1 bis 6)
    # Sucht nach einem Zeilenumbruch, gefolgt vom Start des nächsten Artikel-Headers.
    artikel_pattern = r"\n+(?=^\s*Artikel\s+[1-6]\b)"
    artikel_blocks = re.split(artikel_pattern, text, flags=re.MULTILINE | re.DOTALL)

    artikel_blocks = [a.strip() for a in artikel_blocks if a.strip().startswith("Artikel")]

    result = {}

    # -----------------------
    # 3. PROCESS EACH ARTIKEL BASED ON ITS NUMBER
    # -----------------------
    for artikel in artikel_blocks:
        num_match = re.match(r"Artikel\s+([1-6])\b", artikel)
        if not num_match:
            continue
        nummer = int(num_match.group(1))

        # Entferne den Artikel-Header (erste Zeile) aus dem Text, der analysiert werden soll
        text_without_header = re.sub(r"Artikel\s+[1-6]\b.*?\n", "", artikel, 1, flags=re.DOTALL).strip()


        if nummer == 1:
            # A) ARTIKEL 1: Enthält Paragraphen (§) und Absätze ((1))

            # Splitte zuerst nach Paragraphen (§)
            para_pattern = r"\n+(?=^\s*§\s*\d+\b)"
            paras = re.split(para_pattern, text_without_header, flags=re.MULTILINE | re.DOTALL)

            # Nur echte Paragraphen behalten und trimmen
            paras = [p.strip() for p in paras if p.strip().startswith("§")]

            strukturierter_artikel = {}
            for para_text in paras:
                # Extrahiere Paragraph-Nummer (z.B. "§ 1")
                para_num_match = re.match(r"§\s*(\d+)", para_text)
                if not para_num_match:
                    continue
                para_nummer = int(para_num_match.group(1))

                # Splitte jeden Paragraphen in Absätze ((1), (2), ...)
                absatz_pattern = r"(?=^\s*\(\d+\))"
                absaetze = re.split(absatz_pattern, para_text, flags=re.MULTILINE | re.DOTALL)

                # Der erste Block enthält den §-Header und möglicherweise Absätze davor
                # Wir entfernen den §-Header und trimmen alle Absätze
                absaetze = [a.strip() for a in absaetze]

                # Der erste Absatz beginnt oft ohne (1)
                if not absaetze[0].startswith('('):
                    # Entferne den §-Header aus dem ersten Absatz
                    first_absatz = re.sub(r"§\s*\d+\b.*?\n", "", absaetze[0], 1, flags=re.DOTALL).strip()
                    absaetze[0] = first_absatz

                # Speichere die Absätze unter der Paragraphen-Nummer
                strukturierter_artikel[para_nummer] = [a.strip() for a in absaetze if a.strip()]

            result[nummer] = strukturierter_artikel

        elif nummer in [4, 5]:
            # B) ARTIKEL 4 & 5: Reine Gliederungspunkte (1., 2., 3., ...)

            # Splitte nach dem Muster für Gliederungspunkte
            gliederung_pattern = r"(?=^\s*\d+\.\s+)"
            gliederungspunkte = re.split(gliederung_pattern, text_without_header, flags=re.MULTILINE | re.DOTALL)

            # Der erste Block kann leer sein oder Einleitungstext enthalten
            gliederungspunkte = [g.strip() for g in gliederungspunkte if g.strip()]

            # Wir verwenden die Gliederungspunkte direkt als Liste
            result[nummer] = gliederungspunkte

        else:
            # C) ARTIKEL 2, 3, 6: Keine weitere Aufteilung notwendig
            # Der Text ist der gesamte Block nach dem Header
            result[nummer] = text_without_header

    return result
```

```python
import pandas as pd
import re

def convert_to_dataframes_long_format(structured_data):
    """
    Konvertiert das hierarchische Struktur-Dictionary (Ergebnis von extract_gesetz_structure)
    in drei DataFrames (Long Format) für die verschiedenen Gliederungstypen.
    """

    artikel1_rows = []
    artikel4_5_rows = []
    artikel_rest_rows = []

    for artikel_nummer, inhalt in structured_data.items():

        # A) ARTIKEL 1: Paragraphen und Absätze
        if artikel_nummer == 1 and isinstance(inhalt, dict):
            for para_num, absatz_list in inhalt.items():

                # Gehe davon aus, dass para_num ein Integer ist (aus extract_gesetz_structure)
                for i, absatz_text in enumerate(absatz_list):

                    # Extrahiere Absatznummer (z.B. (1) -> 1)
                    absatz_match = re.match(r"^\((\d+)\)", absatz_text)
                    absatz_nr = int(absatz_match.group(1)) if absatz_match else (i + 1)

                    artikel1_rows.append({
                        "Artikel": artikel_nummer,
                        "Paragraph_Nr": para_num,
                        "Absatz_Nr": absatz_nr,
                        "Text": absatz_text
                    })

        # B) ARTIKEL 4 & 5: Reine Gliederungspunkte (Liste)
        elif artikel_nummer in [4, 5] and isinstance(inhalt, list):
            for i, gliederung_text in enumerate(inhalt):

                # Extrahiere Gliederungsnummer (z.B. 1. -> 1)
                gliederung_match = re.match(r"(\d+)\.", gliederung_text)
                gliederung_nr = int(gliederung_match.group(1)) if gliederung_match else (i + 1)

                artikel4_5_rows.append({
                    "Artikel": artikel_nummer,
                    "Gliederungspunkt_Nr": gliederung_nr,
                    "Text": gliederung_text
                })

        # C) ARTIKEL 2, 3, 6: Gesamter Text
        elif artikel_nummer in [2, 3, 6] and isinstance(inhalt, str):
            artikel_rest_rows.append({
                "Artikel": artikel_nummer,
                "Text": inhalt
            })

    # Erstellung der DataFrames
    df_artikel1 = pd.DataFrame(artikel1_rows)
    df_artikel4_5 = pd.DataFrame(artikel4_5_rows)
    df_artikel_rest = pd.DataFrame(artikel_rest_rows)

    return {
        "Artikel_1_Gliederung": df_artikel1,
        "Artikel_4_5_Gliederung": df_artikel4_5,
        "Artikel_Rest_Text": df_artikel_rest
    }
```

```python
import pandas as pd

def merge_all_dataframes(tabellen_output):
    """
    Führt die drei DataFrames in einem einzigen Master-DataFrame zusammen.
    Fehlende Spalten werden automatisch mit NaN aufgefüllt.
    """

    df1 = tabellen_output["Artikel_1_Gliederung"].copy()
    df4_5 = tabellen_output["Artikel_4_5_Gliederung"].copy()
    df_rest = tabellen_output["Artikel_Rest_Text"].copy()

    # 1. Füge eine Spalte zur Kennzeichnung des Gliederungstyps hinzu
    df1['Typ'] = 'Paragraph/Absatz'
    df4_5['Typ'] = 'Gliederungspunkt'
    df_rest['Typ'] = 'Volltext'

    # 2. Führe die DataFrames vertikal zusammen (Concatenation)
    # Die Funktion concat stellt sicher, dass alle Spaltennamen berücksichtigt werden.
    # Wenn eine Spalte in einem DataFrame fehlt, wird der Wert NaN eingefügt.
    master_df = pd.concat([df1, df4_5, df_rest], ignore_index=True)

    # 3. Bereinige die Spaltenreihenfolge für bessere Lesbarkeit
    # Wir stellen die wichtigen Spalten an den Anfang
    ordered_columns = [
        'Artikel',
        'Typ',
        'Paragraph_Nr',
        'Absatz_Nr',
        'Gliederungspunkt_Nr',
        'Text'
    ]

    # Füge alle anderen Spalten hinzu, falls es welche gibt, die nicht enthalten sind (z.B. durch Fehler)
    final_columns = [col for col in ordered_columns if col in master_df.columns]

    return master_df[final_columns]
```

```python
import pandas as pd
import re

def convert_comments_to_semi_wide_numbered(df_wide_comments):

    # NEU: Entferne die Spalten, die beim Melten neu erstellt werden sollen, falls sie existieren.
    cols_to_drop = ['gliederungspunkt_raw', 'anmerkungstext', 'Merge_Haupt_Nr', 'Merge_Sub_Nr']
    df_wide_comments = df_wide_comments.drop(
        columns=[col for col in cols_to_drop if col in df_wide_comments.columns],
        errors='ignore'
    )

    # rename "artikel6_anmerkungen" to "artikel6_allganmerkungen"
    if 'artikel6_anmerkungen' in df_wide_comments.columns:
        df_wide_comments = df_wide_comments.rename(columns={'artikel6_anmerkungen': 'artikel6_allganmerkungen'})

    # 1. Zuweisung der ID (Mapping)
    org_names = df_wide_comments['organisation_name'].unique()
    org_map = {name: f"Org_{i+1}" for i, name in enumerate(org_names)}

    # 2. Erstelle das Long Format (Melt) und extrahiere die Struktur
    id_vars = ['organisation_name']
    value_vars = [col for col in df_wide_comments.columns if 'anmerkungen' in col]

    # Der Fehler tritt hier auf, da value_name='anmerkungstext' nun garantiert
    # nicht mehr im DataFrame enthalten ist, da es in Schritt 1 entfernt wurde.
    df_long = df_wide_comments.melt(
        id_vars=id_vars,
        value_vars=value_vars,
        var_name='gliederungspunkt_raw',
        value_name='anmerkungstext'
    ).dropna(subset=['anmerkungstext'])

    # ... (Rest der Funktion bleibt wie zuvor)

    # ... [Rest der Funktion zur Extraktion der Struktur und Pivotierung] ...
    pattern = re.compile(
        r"artikel(?P<artikel_nr>\d+)_" +
        r"(?:p(?P<para_nr>\d+)a?(?P<absatz_nr>\d*)|" +
        r"n(?P<norm_nr>\d+)(?:_(?P<sub_nr>\d))?|" +
        r"allg|" +
        r"(?P<art_rest>artikel[2356]|anmerkungen))" +
        r"anmerkungen"
    )

    def extract_structure(raw_key):
        match = pattern.match(raw_key)
        if match:
            artikel = int(match.group('artikel_nr')) if match.group('artikel_nr') else None

            if match.group('para_nr'):
                haupt_nr = int(match.group('para_nr'))
                sub_nr = int(match.group('absatz_nr')) if match.group('absatz_nr') else 1
            elif match.group('norm_nr'):
                haupt_nr = int(match.group('norm_nr'))
                sub_nr = int(match.group('sub_nr')) if match.group('sub_nr') else 0
            else:
                haupt_nr = 0
                sub_nr = 0
                artikel_match = re.search(r"artikel(\d+)_anmerkungen", raw_key)
                if artikel_match: artikel = int(artikel_match.group(1))

            return pd.Series([artikel, haupt_nr, sub_nr, raw_key])
        return pd.Series([None, None, None, None])

    df_long[['Artikel', 'Merge_Haupt_Nr', 'Merge_Sub_Nr', 'Gliederungspunkt_Roh']] = df_long['gliederungspunkt_raw'].apply(extract_structure)

    # 3. PIVOT
    df_semi_wide = df_long.pivot_table(
        index=['Artikel', 'Merge_Haupt_Nr', 'Merge_Sub_Nr', 'Gliederungspunkt_Roh'],
        columns='organisation_name',
        values='anmerkungstext',
        aggfunc='first'
    ).reset_index()

    # 4. Numerische Spaltennamen zuweisen
    new_columns = []
    for col in df_semi_wide.columns:
        if col in org_map:
            new_columns.append(org_map[col])
        else:
            new_columns.append(col)

    df_semi_wide.columns = new_columns
    df_semi_wide.columns.name = None

    df_semi_wide = df_semi_wide.rename(columns={'Gliederungspunkt_Roh': 'Gesetzesteil_ID'})

    df_semi_wide['Artikel'] = df_semi_wide['Artikel'].astype('Int64', errors='ignore')
    df_semi_wide['Merge_Haupt_Nr'] = df_semi_wide['Merge_Haupt_Nr'].astype('Int64', errors='ignore')
    df_semi_wide['Merge_Sub_Nr'] = df_semi_wide['Merge_Sub_Nr'].astype('Int64', errors='ignore')

    return df_semi_wide, org_map
```

```python
import pandas as pd

def merge_final_datasets(master_gesetz_df, df_comments):
    """
    Produce final wide dataset with columns:
    Artikel | Typ | Paragraph | Absatz | Gliederungspunkt_Nr | Text | Org_1 ... Org_N

    Special rule:
    - Allgemeine Anmerkungen get Paragraph = -1, Absatz = 0
      so they do NOT collide with Volltext and the merge is clean.
    """

    # =====================================================
    # 1. PREPARE MASTER (Gesetzestext)
    # =====================================================
    master = master_gesetz_df.copy()

    # Real paragraph number or gliederungspunkt
    master["Paragraph"] = master["Paragraph_Nr"].fillna(master["Gliederungspunkt_Nr"])
    master["Absatz"] = master["Absatz_Nr"].fillna(0)

    # Clean dtypes
    for col in ["Artikel", "Paragraph", "Absatz"]:
        master[col] = pd.to_numeric(master[col], errors="coerce").fillna(0).astype("Int64")

    # =====================================================
    # 2. PREPARE COMMENTS (already semi-wide from previous step)
    # =====================================================
    comments = df_comments.rename(columns={
        "Merge_Haupt_Nr": "Paragraph",
        "Merge_Sub_Nr": "Absatz"
    }).copy()

    for col in ["Artikel", "Paragraph", "Absatz"]:
        comments[col] = pd.to_numeric(comments[col], errors="coerce").fillna(0).astype("Int64")

    # =====================================================
    # 3. APPLY THE NEW RULE:
    #    Allgemeine Anmerkungen = Paragraph -1, Absatz 0
    # =====================================================
    # Condition:
    # Paragraph == 0 and Absatz == 0 AND Gesetzesteil_ID ends with 'anmerkungen'
    is_allg = (
        (comments["Paragraph"] == 0) &
        (comments["Absatz"] == 0) &
        comments["Gesetzesteil_ID"].str.contains(r"anmerkungen$", na=False)
    )

    comments.loc[is_allg, "Paragraph"] = -1
    comments.loc[is_allg, "Absatz"] = 0

    # =====================================================
    # 4. Add synthetic rows to master for Allgemeine Anmerkungen
    #    (otherwise the merge will drop them)
    # =====================================================
    general_articles = comments.loc[is_allg, "Artikel"].unique()

    synthetic_rows = []
    for art in general_articles:
        synthetic_rows.append({
            "Artikel": art,
            "Typ": "Allgemeine Anmerkungen",
            "Paragraph": -1,
            "Absatz": 0,
            "Gliederungspunkt_Nr": None,
            "Text": None,
        })

    if synthetic_rows:
        master = pd.concat([master, pd.DataFrame(synthetic_rows)], ignore_index=True)

    # Ensure types after concat
    for col in ["Artikel", "Paragraph", "Absatz"]:
        master[col] = master[col].astype("Int64")

    # =====================================================
    # 5. LEFT MERGE (keep all Gesetz rows, attach comments)
    # =====================================================
    final = master.merge(
        comments,
        on=["Artikel", "Paragraph", "Absatz"],
        how="left"
    )

    # =====================================================
    # 6. FINAL COLUMN ORDER
    # =====================================================
    org_cols = [c for c in final.columns if c.startswith("Org_")]

    final = final[
        ["Artikel", "Typ", "Paragraph", "Absatz", "Gliederungspunkt_Nr", "Text"]
        + org_cols
    ]

    return final
```

```python
import pandas as pd
import re

def export_to_excel(final_df, org_map, filename="gesetz_kommentare.xlsx", na_value=""):
    """
    Exports:
      - final_df     → sheet 'Gesetz + Kommentare'
      - org_map      → sheet 'Org_Map'
    with:
      - consistent NA handling
      - sorted organisation columns
      - sorted rows by Artikel, Typ, Paragraph, Absatz, Gliederungspunkt_Nr
    """

    # =======================================================
    # 1. SORT ORGANISATION COLUMNS (Org_1, Org_2, Org_3 …)
    # =======================================================
    org_cols = [c for c in final_df.columns if c.startswith("Org_")]

    # Sort numerically by the number after "Org_"
    org_cols_sorted = sorted(org_cols, key=lambda x: int(re.findall(r'\d+', x)[0]))

    # Main identifying columns
    id_cols = ["Artikel", "Typ", "Paragraph", "Absatz", "Gliederungspunkt_Nr", "Text"]

    # Rebuild dataframe with sorted organization columns
    export_df = final_df[id_cols + org_cols_sorted].copy()

    # =======================================================
    # 2. SORT ROWS
    # =======================================================
    # Ensure sorting columns exist
    for col in ["Artikel", "Typ", "Paragraph", "Absatz", "Gliederungspunkt_Nr"]:
        if col not in export_df.columns:
            export_df[col] = None

    export_df = export_df.sort_values(
        by=["Artikel", "Typ", "Paragraph", "Absatz", "Gliederungspunkt_Nr"],
        ascending=[True, True, True, True, True]
    )

    # =======================================================
    # 3. Fill NA values consistently
    # =======================================================
    export_df = export_df.fillna(na_value)

    # =======================================================
    # 4. Build Org Map table and sort
    # =======================================================
    org_map_df = pd.DataFrame(
        [{"Org_ID": v, "Organisation_Name": k} for k, v in org_map.items()]
    )

    # Sort by numeric value in Org_X
    org_map_df = org_map_df.sort_values(
        by="Org_ID",
        key=lambda col: col.apply(lambda x: int(re.findall(r'\d+', x)[0]))
    )

    # Replace NA values in org map too
    org_map_df = org_map_df.fillna(na_value)

    # =======================================================
    # 5. Write Excel file with two sheets
    # =======================================================
    with pd.ExcelWriter(filename, engine="openpyxl") as writer:
        export_df.to_excel(writer, sheet_name="Gesetz + Kommentare", index=False)
        org_map_df.to_excel(writer, sheet_name="Org_Map", index=False)

    print(f"Excel export completed: {filename}")
```

```python
df_long = final_combined_df.melt(
    id_vars=["Artikel", "Typ", "Paragraph", "Absatz", "Gliederungspunkt_Nr", "Text"],
    var_name="Organisation",
    value_name="Comment"
).dropna(subset=["Comment"])
```

```python
df_long
```

```python
# sort by Artikel, Paragraph, Absatz, Gliederungspunkt_Nr
df_long = df_long.sort_values(
    by=["Artikel", "Paragraph", "Absatz", "Gliederungspunkt_Nr"],
    ascending=[True, True, True, True]
)
```

```python
df_long.head(20)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Clean text fields
df_long["Text"] = df_long["Text"].fillna("").astype(str)
df_long["Comment"] = df_long["Comment"].fillna("").astype(str)

# Optionally filter out empty comments
df_long = df_long[df_long["Comment"].str.strip() != ""]

# Encode safely
df_long["text_emb"] = df_long["Text"].apply(
    lambda x: model.encode(x, convert_to_tensor=True)
)
df_long["comment_emb"] = df_long["Comment"].apply(
    lambda x: model.encode(x, convert_to_tensor=True)
)

df_long["similarity"] = df_long.apply(
    lambda r: float(util.cos_sim(r["text_emb"], r["comment_emb"])), axis=1
)
```

```python
df_long
```

```python
section_sim = df_long.groupby("Artikel")["similarity"].mean()

section_sim
```
